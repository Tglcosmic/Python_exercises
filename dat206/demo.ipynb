```python
import matplotlib.pyplot as plt
import numpy as np

```

```python

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

#Tùy chỉnh đồ thị
plt.plot(x, y, label = 'sin(x)')

#tùy chỉnh đồ thị
plt.title('Biểu đồ đường sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()

# hiển thị đồ thị
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt

# Tạo dữ liệu ngẫu nhiên
data = np.random.randn(1000)

# Tạo histogram
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, color='skyblue', alpha=0.7)  # Thay đổi màu sắc ở đây

# Thêm tiêu đề và nhãn cho biểu đồ
plt.title('Biểu đồ Histogram')
plt.xlabel('Giá trị')
plt.ylabel('Tần suất')

plt.grid(True)
plt.show()
```
