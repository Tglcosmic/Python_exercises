```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
filepath = r"H:\Python_exercises\dat206\data\BankRateby12.csv"
data = pd.read_csv(filepath)
df = pd.DataFrame(data)

bank = df['Bank']
interest_rate = df['Interest_Rate']

plt.figure(figsize=(6, 4))
total = data['Interest_Rate'].sum()
plt.pie(interest_rate, labels=bank, autopct=lambda p:'{:0.1f}%'.format(p*total/100), startangle=90, colors= np.random.rand(10, 4), pctdistance=.7, labeldistance=1.1)

plt.title("Biu  th hin li sut ca cc ngn hng")

plt.axis("equal")
plt.tight_layout()
plt.show()
```
