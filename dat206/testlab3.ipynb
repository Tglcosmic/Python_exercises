```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

```

```python
# Dữ liệu mẫu: Điểm thi của học sinh 
exam_scores = np.random.normal(0, 1, 1000)
#Ve histogram
plt.hist(exam_scores, bins=30, color='skyblue', \
edgecolor='darkblue', alpha=0.7)
# Cấu hình đô thị
plt.title('Histogram của Điểm Thi')
plt.xlabel('Điểm Thi')
plt.ylabel('Số Lượng Học Sinh ')
# Hiện thị đô thị
plt.show()
```
