# bai 1

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Dữ liệu sản lượng sản xuất hằng tháng
months = np.arange(1, 13) # Từ tháng 1 đến tháng 12
production = np.random.randint(1000,5000,size=12)

plt.plot(months, production, color="b", linestyle="-", marker="o")

plt.xlabel('month')
plt.ylabel('production')
plt.title('Bieu do san luong')

plt.grid()
plt.xticks()
plt.tight_layout()
plt.show
```

```python
quater = ['quý 1', 'quý 2', 'quý 3', 'quý 4']
sales = [1000, 2000, 3000, 7000]

plt.bar(quater ,sales, color="g")

plt.xlabel('quý')
plt.ylabel('doanh số bán hàng')
plt.title('Biểu đồ doanh số bán hàng')

# for i in range(len(quater)):
#     plt.text(quater[i], sales[i], str(sales[i]), ha='center', va='bottom') 
plt.grid(True)
plt.tight_layout()
plt.show
```

# bài 3


```python
categories = ['Áo', 'Quần', 'Giày', 'Phụ kiện']
sales = np.random.randint(50000, 150000, size=len(categories))


plt.figure(figsize=(4, 4))
plt.pie(sales, labels=categories, autopct='%1.1f%%', startangle=90)

plt.title('Biểu đồ thể hiện phần trăm doanh số bán hàng theo sản phẩm')

plt.axis('equal')
plt.tight_layout()
plt.show()
```
