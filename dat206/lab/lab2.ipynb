```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Bài 1

```python
# Dữ liệu mẫu
ngay = ['1/1', '1/2', '1/3', '1/4', '1/5', '1/6']
ca_mac = [100, 150, 200, 180, 220, 250]
ca_tu_vong = [5, 8, 12, 10, 15, 18]
ca_hoi_phuc = [40, 60, 80, 70, 90, 100]
color = np.random.rand(3)

# Vẽ biểu đồ đường
plt.plot(ngay, ca_mac, ls='-', c= color, label='ca mắc')
plt.plot(ngay, ca_tu_vong, ls='-.', c= color, label='ca tử vong')
plt.plot(ngay, ca_hoi_phuc, ls='--', c= color, label= 'ca hồi phục')

# Thêm chú thích và tiêu đề plt.title('Tình hình COVID-19') plt.xlabel('Ngày ')
plt.xlabel('Ngày')
plt.ylabel('Số lượng')
plt.title('Tình hình COVID19')
plt.legend(loc = 'upper right', title= "ghi chú")
# Hiển thị biểu đồ
plt.grid(True)
plt.show()
```

# Bài 2


```python
# Dữ liệu mẫu
san_pham = [ 'Điện thoại', 'Máy tính', 'Tivi', 'Đồ gia dụng']
doanh_so = [200, 300, 250, 150] # Doanh số bán hàng (đơn vị: triệu VND)

# Ký hiệu cho mỗi sản phẩm
ky_hieu = ['DT', 'MT', 'TV', 'ĐGD']

# Tạo mảng màu sắc đa dạng 
colors = np.random.rand(10, 3)

# Vẽ biểu đồ cột
for i, sp in enumerate(san_pham): 
    plt.bar(san_pham[i], doanh_so[i], color=colors[i], label=ky_hieu[i])
# Chú thích và tiêu đề
plt.title("Doanh số bán hàng của các sản phẩm")
plt.xlabel("Sản phẩm")
plt.ylabel("Doanh số bán hàng (triệu VNĐ)") 
plt.legend(title="Ký hiệu")
# Hiện thị biểu đồ
plt.grid(axis='y', linestyle='--', alpha=0.7) 
plt.tight_layout()
plt.show()
```

# Bài 3


```python

# Dữ Liệu mẫu
san_pham = ['Sản phẩm A', 'Sản phẩm B', 'Sản phẩm C']
phan_tram_doanh_so = [30, 40, 38] # Phân trăm doanh số bán hàng của các sản phẩm trong tháng

# Tạo một mảng màu sắc đa dạng cho từng sản phẩm
colors= plt.cm.tab10(np.linspace(0, 1, len(san_pham)))
# Vẽ biểu đồ tròn
plt.figure(figsize=(8, 6))
patches, texts, autotexts = plt.pie(phan_tram_doanh_so, labels=san_pham, autopct="%1.1f%%", startangle=90, colors=colors)

# Định dạng chữ cho chú thích
for text, autotext in zip(texts, autotexts):
    text.set_fontsize(12) 
    autotext.set_fontsize(12)

# Đặt tiêu đề cho biểu đồ
plt.title("Phân tích doanh số bán hàng của các sản phẩm trong một tháng", fontsize=16)
# Chú thích 
plt.legend(loc= 'center')
# Hiện thị biều đồ 
plt.axis('equal') 
plt.tight_layout() 
plt.show()
```

# Bài 4


```python
filepath = r"H:\Python_exercises\dat206\data\timeseries_stockprice.csv"
df = pd.read_csv(filepath)
data = pd.DataFrame(df)
date = data['Date']
Amazon = data['Amazon']
Google = data['Google']
Facebook = data['Facebook']
    
barwidth = 0.25     # Độ rộng của bar
bar1 = np.arange(len(date))     # tính chiều dài của Date để set vị trí đầu tiên cho cột
bar2 = [i + barwidth for i in bar1]
bar3 = [i + barwidth for i in bar2]

plt.figure(figsize=(20, 5))
# plt.plot(date, Amazon, ls='--', c='r', label='Amazon')
# plt.plot(date, Google, ls='-.', c='b', label='Google')
# plt.plot(date, Facebook, ls=':', c='g', label='Facebook')
barplot1 = plt.bar(bar1, Amazon, color='tomato', width=barwidth, align='center', label='Amazon', hatch='.', edgecolor='black')
plt.bar_label(barplot1, labels=data['Amazon'], padding=3, rotation=90)
barplot2 = plt.bar(bar2, Google, color='cyan', width=barwidth, align='center', label='Google', hatch='//', edgecolor='black')
plt.bar_label(barplot2, labels=data['Google'], padding=3, rotation=90)
barplot3 = plt.bar(bar3, Facebook, color='lime', width=barwidth, align='center', label='Facebook', hatch='*', edgecolor='black')
plt.bar_label(barplot3, labels=data['Facebook'], padding=3, rotation=90)


plt.title('Timeseries stockprice')
plt.xlabel('Date')
plt.ylabel('stockprice')
plt.xticks(bar1, data['Date'], rotation=45)

plt.legend(loc='upper right', title='Chú thích', fontsize='10', fancybox=True, shadow=True)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Vẽ biểu đồ
plt.figure(figsize=(20, 5))
plt.plot(date, Amazon, color="b", linestyle="-", marker="o", label= 'Amazon')
plt.plot(date, Facebook, color="g", linestyle="--", marker="o", label= 'Facebook')
plt.plot(date, Google, color="r", linestyle="-.", marker="o", label= 'Google')

plt.xlabel('month')
plt.ylabel('stock price')
plt.title('Biểu đồ biểu diễn giá trị cổ phiếu')

plt.legend()
plt.grid()
plt.xticks()
plt.tight_layout()
plt.show

```
