```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
# Sample high school exam scores
exam_scores = [65, 70, 72, 75, 78, 80, 82, 85, 88, 90, 92, 95, 98, 100]

# Create a histogram
plt.figure(figsize=(14, 6))
plt.hist(exam_scores, bins=10, edgecolor='black', alpha=0.7)
plt.title('High School Exam Scores Distribution')
plt.xlabel('Exam Scores')
plt.ylabel('Number of Students') 
plt.grid(True)

# Calculate statistics
mean_score = np.mean(exam_scores)
median_score = np.median(exam_scores)
std_dev = np.std(exam_scores)

# Add statistics to the plot
plt.axvline(mean_score, color='r', linestyle='dashed', linewidth=1, label=f'Mean: {mean_score:.2f}')
plt.axvline(median_score, color='g', linestyle='dashed', linewidth=1, label=f'Median: {median_score:.2f}')
plt.axvline(mean_score - std_dev, color='b', linestyle='dashed', linewidth=1, label=f'Standard Deviation: {std_dev:.2f}')
plt.axvline(mean_score + std_dev, color='b', linestyle='dashed', linewidth=1)

# Add text annotations
plt.text(mean_score + 1, plt.ylim()[1] * 0.9, f'Mean: {mean_score:.2f}', color='r')
plt.text(median_score + 1, plt.ylim()[1] * 0.85, f'Median: {median_score:.2f}', color='g')
plt.text(mean_score + std_dev + 1, plt.ylim()[1] * 0.8, f'Standard Deviation: {std_dev:.2f}', color='b')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

# Bai 2

```python
# Sample salary data
salary_data = {
    'Department': ['HR', 'HR', 'Finance', 'Finance', 'IT', 'IT', 'IT', 'Marketing', 'Marketing'], 
    'Salary': [45800, 50000, 60000, 65000, 55080, 60000, 70000, 48000, 52000]
}

# Convert data to DataFrame
df = pd.DataFrame(salary_data)

# Basic boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Salary'], color='g')
plt.title('Salary Distribution in the Company')
plt.xlabel('Salary ($)')
plt.grid(True)
plt.show()

# Advanced boxplot with department comparison
plt.figure(figsize=(10, 6))
sns.boxplot(x='Department', y='Salary', data=df)
plt.title('Salary Distribution by Department')
plt.xlabel('Department')
plt.ylabel('Salary ($)')
plt.grid(True)
plt.show()
```

```python
# Generate fake age data
np.random.seed(0)
age_male = np.random.normal(loc=30, scale=10, size=500)
age_female = np.random.normal(loc=28, scale=8, size=500)

# Basic density plot
plt.figure(figsize=(10, 6))
sns.kdeplot(age_male, label='Male', shade=True)
sns.kdeplot(age_female, label='Female', shade=True)
plt.title('Density Plot of Age Distribution by Gender')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Táº¡o dá»¯ liá»u giáº£ Äá»nh vá» giÃ¡ cá» phiáº¿u
np.random.seed(0)
dates = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')
prices = 100 + np.cumsum(np.random.normal(loc=0, scale=1, size=len(dates)))

# Táº¡o DataFrame tá»« dá»¯ liá»u giáº£ Äá»nh
stock_data = pd.DataFrame({'Date': dates, 'Price': prices})

# Biá»u Äá» time plot cÆ¡ báº£n
plt.figure(figsize=(10, 6))
plt.plot(stock_data['Date'], stock_data['Price'], label='Stock Price')
plt.title('Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()

# TÃ­nh toÃ¡n vÃ  váº½ trend Line
X = stock_data['Date'].astype('int64').values.reshape(-1, 1)
y = stock_data['Price'].values
model = LinearRegression().fit(X, y)
trend_line = model.predict(X)

# Biá»u Äá» trend line
plt.plot(stock_data['Date'], trend_line, label='Trend Line', color='red')
plt.legend()
plt.show()
```
