```python
import pandas as pd
import numpy as np
```

```python
filepath = r"data\lab2.csv"
data = pd.read_csv(filepath)
df = pd.DataFrame(data)
```

## Bài 1


#### Tính tổng điểm của tất cả học sinh theo giới tính.

```python
df.groupby('gender')['score'].sum().reset_index()
```

#### Tính điểm trung bình của các học sinh theo nhóm tuổi.

```python
df['type_age'] = df['age'].apply(lambda x: '18 => 20' if x <= 20 else '> 20')                                                              
df
```

```python
df_grouped = df.groupby('type_age')['score'].mean().reset_index()
df_grouped
```

#### Tính số lượng học sinh có điểm số trên 80 theo tên.


```python
df[df['score'] > 80].groupby('name').size().reset_index(name='Số lượng học sinh')
```

## Bài 2

#### Nhóm học sinh theo giới tính và sau đó tính tổng điểm của từng nhóm.


```python
df.groupby('gender')['score'].sum()
```

#### Nhóm học sinh trong mỗi nhóm giới tính theo nhóm tuổi và tính tổng điểm của từng nhóm con.

```python
def age_group(age):
    if age < 25:
        return '< 25'
    else:
        return '> 25'
df['age_group'] = df['age'].apply(age_group)
grouped = df.groupby(['gender','age_group'])
total_scores = grouped['score'].sum().reset_index()
total_scores
```

#### Nhóm học sinh trong mỗi nhóm giới tính theo nhóm tuổi và tính điểm trung bình của từng nhóm con.

```python
grouped = df.groupby(['gender', 'age'])
average_scores = grouped['score'].mean().reset_index()
average_scores
```

## Bài 3


#### Tính tổng điểm và tuổi trung bình của học sinh có điểm số lớn hơn 80 theo giới tính

```python
filtered_df = df[df['score'] > 80]

grouped = filtered_df.groupby('gender')

result = grouped.agg(
    Tổng_Điểm=('score', 'sum'),
    Tuổi_Trung_Bình=('age', 'median')
).reset_index()
result
```

#### Tính số lượng học sinh có điểm số nằm trong khoảng từ 70 đến 90 theo nhóm tuổi và giới tính.

```python
filtered_df = df[(df['score'] >= 70) & (df['score'] <= 90)]
print(filtered_df)

grouped = filtered_df.groupby(['gender', 'age_group'])

count_students = grouped.size().reset_index(name='Số lượng học sinh')
count_students
```
