```python
import random
import pandas as pd
import numpy as np
from datetime import datetime

names = ['Tony Tèo', 'Khanhvc', 'Khanh Đẹp Trai', 'Alice', 'Bob', 'Charlie', 'David', 'Eva']
rows = 10

data = [{
    'Name': random.choice(names)
    ,'YOB': random.randint(1970, 2000)
} for _ in range(rows)]

df = pd.DataFrame(data)

# Tính tuổi từ năm sinh

# current_year = pd.Timestamp.now().year  -cách 2 k import datetime
df['Age'] = df['YOB'].apply(lambda x: datetime.now().year - x) # có thể thêm datetime.now().year thẳng vào hàm
print(df)
```

#### các bạn tìm hiểu cách lấy phần hồng hồng của 2 df ra giúp thầy (dữ liệu tụ chọn) trong zalo

```python
import pandas as pd

# Tạo hai DataFrame mẫu
df1 = pd.DataFrame({
    'key': ['A', 'B', 'C'],
    'value1': [1, 2, 3]
})

df2 = pd.DataFrame({
    'key': ['A', 'B', 'D'],
    'value2': [4, 5, 6]
})

# Hợp nhất hai DataFrame, giữ lại tất cả các hàng từ df2 (bên phải)
df_merged = pd.merge(df1, df2, on='key', how='right')

# Chỉ giữ lại các cột từ df1
result = df_merged[['key', 'value1']]

# Loại bỏ các hàng có giá trị NaN trong cột 'value1'
result_clean = result.dropna()

print("Kết quả hợp nhất và lọc:")
print(result_clean)
```
