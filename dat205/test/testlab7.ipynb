```python
import pandas as pd
from faker import Faker
import random
departments = ['HR', 'Marketing', 'Sales', 'IT', 'Finance']

rows = 100
start_date='-5y'
end_date='today'

fake = Faker()
data = [{
     'Name': fake.name()
    ,'Age': random.randint(18, 65)
    ,'Salary': random.randint(30, 100) * 100
    ,'Department': random.choice(departments)
    ,'JoiningDate': fake.date_between(start_date=start_date, end_date=end_date) 
   
} for _ in range(rows)]
df = pd.DataFrame(data)
df
```

```python
ls_cols = ['Name', 'Age', 'Salary', 'JoiningDate']
df[ls_cols].to_csv('071_suprisemotherfucker.csv'
          , index= False
          , header=True
          , encoding='utf-8'
          , sep=','
          )
```

```python
df_read = pd.read_csv('071_suprisemotherfucker.csv'
                      , encoding='UTF-8'
                      , parse_dates=['JoiningDate']
                      )
print(df_read)
df_read.head()
```
