####  Tính tổng số ca nhiễm, tử vong, và khỏi bệnh trên toàn thế giới.

```python
print(f'Tổng số ca nhiễm toàn thế giới: {df['Confirmed'].sum()}')
print(f'Tổng số ca nhiễm toàn thế giới: {df['Deaths'].sum()}')
print(f'Tổng số ca nhiễm toàn thế giới: {df['Recovered'].sum()}')
```

####  Tính tỷ lệ tử vong trên số ca nhiễm

```python
print(f'tỷ lệ tử vong trên số ca nhiễm:\n{df['Deaths'] / df['Confirmed']}')
```
